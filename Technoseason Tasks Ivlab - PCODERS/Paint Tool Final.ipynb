```python
# Paint tool with Trackbars

# Imorting required libraries

import cv2
import numpy as np

draw = False
window_title = "Paint tool Application"
color_position = [(400, 30), (490,90)]
colour_track = {'Blue': 0, 'Green': 0, 'Red': 0}

img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow(window_title)

# Initial color window, showing black
cv2.rectangle(img, color_position[0], color_position[1], (0,0,0), -1)

#Allocating position to the colour name on the application window

font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.putText(img, "Red: ", (10, 30), font, 0.5, (255,255,255), 1)
img = cv2.putText(img, "Green: ", (90, 30), font, 0.5, (255,255,255), 1)
img = cv2.putText(img, "Blue: ", (170, 30), font, 0.5, (255,255,255), 1)

img = cv2.putText(img, "0", (60, 30), font, 0.5, (255,255,255), 1)
img = cv2.putText(img, "0", (140, 30), font, 0.5, (255,255,255), 1)
img = cv2.putText(img, "0", (220, 30), font, 0.5, (255,255,255), 1)

# Defining functions

def none(x):#for brush size
    pass

def change_R_value(x):#for red colour
    global font, img, colour_track
    img = cv2.putText(img, f"{colour_track['Red']}", (60, 30), font, 0.5, (0,0,0), 1)
    img = cv2.putText(img, f"{x}", (60, 30), font, 0.5, (255,255,255), 1)
    colour_track['Red'] = x

def change_G_value(x):#for green colour
    global font, img, colour_track
    img = cv2.putText(img, f"{colour_track['Green']}", (140, 30), font, 0.5, (0,0,0), 1)
    img = cv2.putText(img, f"{x}", (140, 30), font, 0.5, (255,255,255), 1)
    colour_track['Green'] = x

def change_B_value(x):#for blue colour
    global font, img, colour_track
    img = cv2.putText(img, f"{colour_track['Blue']}", (220, 30), font, 0.5, (0,0,0), 1)
    img = cv2.putText(img, f"{x}", (220, 30), font, 0.5, (255,255,255), 1)
    colour_track['Blue'] = x

def draw_circle(event, x, y, flags, param):#for continuing the process of drawing
    global draw, img

    if event == cv2.EVENT_LBUTTONDOWN:
        draw = True

    elif event == cv2.EVENT_MOUSEMOVE:
        if draw:
            cv2.circle(img, (x,y), cv2.getTrackbarPos("Brush Size", window_title),
                       (cv2.getTrackbarPos("Blue", window_title),
                        cv2.getTrackbarPos("Green", window_title),
                        cv2.getTrackbarPos("Red", window_title)),
                       -1)

    elif event==cv2.EVENT_LBUTTONUP:
        draw = False
        cv2.circle(img, (x,y), cv2.getTrackbarPos("Brush Size", window_title),
                       (cv2.getTrackbarPos("Blue", window_title),
                        cv2.getTrackbarPos("Green", window_title),
                        cv2.getTrackbarPos("Red", window_title)),
                       -1)
        
# Creating Trackbars for colour values

cv2.createTrackbar("Red", window_title, 0 ,255, change_R_value)
cv2.createTrackbar("Green", window_title, 0, 255, change_G_value)
cv2.createTrackbar("Blue", window_title, 0, 255, change_B_value)
cv2.createTrackbar("Brush Size", window_title, 0, 8, none)
cv2.setMouseCallback(window_title, draw_circle)

while(1):
    cv2.imshow(window_title, img)
    key = cv2.waitKey(1) & 0xff
    if key==ord('q'):
        break

    b = cv2.getTrackbarPos("Blue", window_title)
    g = cv2.getTrackbarPos("Green", window_title)
    r = cv2.getTrackbarPos("Red", window_title)
    cv2.rectangle(img, color_position[0], color_position[1], (b,g,r), -1)

cv2.destroyAllWindows()
```
