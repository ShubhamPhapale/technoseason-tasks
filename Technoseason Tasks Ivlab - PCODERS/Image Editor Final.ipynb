```python
#Importing required libraries
import numpy as np
import matplotlib.pyplot as plt

#Reading images
#Use same size images for blending

img = plt.imread("Downloads/1.webp")
img2 = plt.imread("Downloads/Python Logo.jpg")

#Defining functions

def crop(image,d1,d2,d3,d4):#for cropping
    (h,w)=image.shape[:2]
    imgcrop = image[d4:h-d2,d1:w-d3]
    return imgcrop

def blending(image1,image2,weight1,weight2,gamma):#for blending
    bimage = np.ones(image2.shape, dtype="uint8")
    #iterate through rows
    for i in range(len(image1)):
        # iterate through columns
        for j in range(len(image1[0])):
            bimage[i][j] = (image1[i][j]*weight1 + image2[i][j]*weight2 + gamma).astype(np.uint8)
    return bimage

def bright_manipulate(image,brightness):#for changing brightness
    bimage = np.ones(image.shape, dtype="uint8")
    #iterate through rows
    for i in range(len(image)):
        # iterate through columns
        for j in range(len(image[0])):
            bimage[i][j] = brightness/100 + image[i][j]
    return bimage

def rotate(image,direction,angle):#for ratating image

    h,w,c = image.shape
    newimage=np.zeros([h,w,c],dtype=np.uint8)
    if(direction=="Anticlockwise" and angle == 90):
        for i in range(h):
            for j in range(w):
                newimage[i,j]=image[j-1,w-i-1]
                newimage=newimage[0:h,0:w]

    if (direction == "Clockwise" and angle == 90):
        for i in range(h):
            for j in range(w):
                newimage[i,j]=image[h-j-1,w-i-1]
                newimage=newimage[0:h,0:w]

    if (angle==180):
        for i in range(h):
            for j in range(w):
                newimage[i, j] = image[h - i - 1, w - j - 1]
                newimage = newimage[0:h, 0:w]
    return newimage
flag=1

#Using loop for continuing editing

while(flag==1):
    print("1: Crop image")
    print("2: Blend 2 images")
    print("3: Change image Brightness")
    print("4: Rotate image")
    print("5: Exit")
    choice=int(input("Enter your choice "))
    if(choice==1):
        d1 = int(input("Enter distance from left edge "))
        d2 = int(input("Enter distance from bottom edge "))
        d3 = int(input("Enter distance from right edge "))
        d4 = int(input("Enter distance from top edge "))
        showimage = crop(img,d1,d2,d3,d4)
        fig = plt.figure()
        ax = fig.subplots()
        ax.imshow(showimage)
        ax.axis('off')
        plt.show()
        
    else:
        print("Editing Completed")
        
        

    if(choice==2):
        w1=float(input("Enter weight of 1st image "))
        w2=1-w1
        g=float(input("Enter gamma value(0-1) "))
        showimage = blending(img,img2,w1,w2,g)
        fig = plt.figure()
        ax = fig.subplots()
        ax.imshow(showimage)
        ax.axis('off')
        plt.show()
        
    if(choice==3):
        bright=int(input("Enter the percentage change in brightness(-50 to 50) "))
        showimage = bright_manipulate(img,bright)
        fig = plt.figure()
        ax = fig.subplots()
        ax.imshow(showimage)
        ax.axis('off')
        plt.show()
        
    if(choice==4):
        dir=input("Enter direction ")
        angle=int(input("Enter angle "))
        showimage=rotate(img,dir,angle)
        fig = plt.figure()
        ax = fig.subplots()
        ax.imshow(showimage)
        ax.axis('off')
        plt.show()
        
    if(choice==5):
        flag=0
```
