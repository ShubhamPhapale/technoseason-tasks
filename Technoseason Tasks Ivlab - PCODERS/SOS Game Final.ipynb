```python
s1=0 #Score of Player 1
s2=0 #Score of Player 2

size=int(input("Enter size of grid "))
grid=[]

c1=input("Enter character of player 1 (in Capital) ")
c2=input("Enter character of player 2 (in Capital)")

for i in range(size):
    grid.append(['0']*size)



def hcheck(arr,size,a,b): #To check combination in rows
    global s1
    global s2
    for row in range(0,size):
        for column in range(0,size-2):
            if ( arr[row][column]=='S' and arr[row][column+1]==a and arr[row][column+2]=='S'):
                s1=s1+10
            if( arr[row][column]=='S' and arr[row][column+1]==b and arr[row][column+2]=='S'):
                s2=s2+10

def vcheck(arr,size,a,b): #To check combination in columns
    global s1
    global s2
    for row in range(0,size):
        for column in range(0,size-2):
            if ( arr[column][row]=='S' and arr[column+1][row]==a and arr[column+2][row]=='S'):
                s1=s1+10
            if( arr[column][row]=='S' and arr[column+1][row]==b and arr[column+2][row]=='S'):
                s2=s2+10

def dcheck(arr,size,a,b): #To check combination in diagonals
    global s1
    global s2
    for i in range(size):
        for j in range(size):
            w=i-1
            x=j+1
            y=i-2
            z=j++2
            if(w<size and x<size and y<size and z<size):
                if(arr[i][j]=='S' and arr[w][x]==a and arr[y][z]=='S'):
                    s1=s1+10
                if (arr[i][j] == 'S' and arr[w][x] == b and arr[y][z] == 'S'):
                    s2 = s2 + 10

    for i in range(size):
        for j in range(size):
            w=i+1
            x=j+1
            y=i+2
            z=j++2
            if(w<size and x<size and y<size and z<size):
                if(arr[i][j]=='S' and arr[w][x]==a and arr[y][z]=='S'):
                    s1=s1+10
                if (arr[i][j] == 'S' and arr[w][x] == b and arr[y][z] == 'S'):
                    s2 = s2 + 10

def showgrid(arr,size): #To show the current filled grid
    for i in range(size):
            print(grid[i])

def compgrid(arr,size): #To check if every position is filled
    flag=0
    for i in range(size):
        for j in range(size):
            if (arr[i][j] == '0'):
                flag = 1
    if(flag==1):
        return 0
    else:
        return 1
def writegrid(arr,row,colm,character): #To enter character in grid
    arr[row][colm]=character

#Use of loop to continue the game

while(compgrid(grid,size)==0):
    print("Player 1: Enter Location")
    r=int(input("Enter row number "))
    c=int(input("Enter column number "))
    char=input("Enter either 'S' or your character ")
    writegrid(grid,r,c,char)
    showgrid(grid,size)
    if (compgrid(grid,size)==1):
        break
    print("Player 2: Enter Location")
    r = int(input("Enter row number "))
    c = int(input("Enter column number "))
    char = input("Enter either 'S' or your character ")
    writegrid(grid, r, c, char)
    showgrid(grid, size)
    if (compgrid(grid, size) == 1):
        break

hcheck(grid,size,c1,c2)
vcheck(grid,size,c1,c2)
dcheck(grid,size,c1,c2)
print("Player 1 score : ",s1)
print("Player 2 score : ",s2)

if(s1>s2):
    print("Player 1 wins")
elif(s1==s2):
    print("It is a tie")
else:
    print("Player 2 wins")
```
